# **Bài 1. Kiểm tra học lực sinh viên (if – else)**

Yêu cầu:

Nhập điểm trung bình (GPA) của một sinh viên (từ 0 đến 10) và phân loại học lực:

GPA ≥ 8.5 → Xuất sắc

7.0 ≤ GPA < 8.5 → Giỏi

5.5 ≤ GPA < 7.0 → Khá

4.0 ≤ GPA < 5.5 → Trung bình

GPA < 4.0 → Yếu

```python
gpa = int(input("Nhập điểm trung bình (GPA): "))
if gpa >= 8.5:
  print("Giỏi")
elif gpa >= 5.5:
  print("Khá")
elif gpa >= 4.0:
  print("Trung bình")
else:
  print("Yếu")
```

# **Bài 2. Tính tổng các số chẵn trong khoảng (for)**

Yêu cầu:
Nhập một số nguyên dương n.

Tính tổng các số chẵn từ 1 đến n.

```python
so = int(input("Nhập số nguyên dương n: "))
tong = 0
for i in range(1, so + 1):
    if i % 2 == 0:
        tong += i
print(f"Tổng các số chẵn từ 1 đến {so} là: {tong}")
```

# **Bài 3. Đếm số lần nhập đúng mật khẩu (while + if)**

Yêu cầu:

Mật khẩu đúng là "python123"

Cho phép người dùng nhập mật khẩu tối đa 3 lần

Nếu nhập đúng → in "Đăng nhập thành công"

Nếu nhập sai quá 3 lần → in "Tài khoản bị khóa"



```python
so_lan = int(input("Nhập số lần nhập: "))
if so_lan > 3:
  print("Tài khoản bị khóa")
while so_lan <= 3:
  mat_khau = input("Nhập mật khẩu: ")
  if mat_khau == ("python123"):
    print("Đăng nhập thành công")
    break
  if mat_khau != ("python123"):
    print("Mật khẩu không đúng")
    so_lan += 1
    if so_lan > 3:
      print("Tài khoản bị khóa")
    break
```
